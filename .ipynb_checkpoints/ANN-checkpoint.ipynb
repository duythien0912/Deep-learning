```python
import os
import random
import skimage.data
import skimage.transform
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf

ROOT_PATH = "./traffic"
# 4591 images for training
train_data_dir = os.path.join(ROOT_PATH, "datasets/BelgiumTS/Training")
# 2534 images for testing
test_data_dir = os.path.join(ROOT_PATH, "datasets/BelgiumTS/Testing")

def load_data(data_dir):
    directories = [d for d in os.listdir(data_dir) 
                   if os.path.isdir(os.path.join(data_dir, d))]
    labels = []
    images = []
    for d in directories:
        label_dir = os.path.join(data_dir, d)
        file_names = [os.path.join(label_dir, f) 
                      for f in os.listdir(label_dir) if f.endswith(".ppm")]
        for f in file_names:
            images.append(skimage.data.imread(f))
            labels.append(int(d))
    return images, labels

images, labels = load_data(train_data_dir)

def display_label_images(images, label):
    limit = 24  # show a max of 24 images
    plt.figure(figsize=(15, 5))
    i = 1

    start = labels.index(label)
    end = start + labels.count(label)
    for image in images[start:end][:limit]:
        plt.subplot(3, 8, i)  # 3 rows, 8 per row
        plt.axis('off')
        i += 1
        plt.imshow(image)
    plt.show()

display_label_images(images, 32)

```

```python
for image in images[:5]:
    print("shape: {0}, min: {1}, max: {2}".format(image.shape, image.min(), image.max()))
```

```python

images32 = [skimage.transform.resize(image, (32, 32), mode='constant')
                for image in images]

def display_images_and_labels(images, labels):
    unique_labels = set(labels)
    plt.figure(figsize=(15, 15))
    i = 1
    for label in unique_labels:
        # Pick the first image for each label.
        image = images[labels.index(label)]
        plt.subplot(8, 8, i)  # A grid of 8 rows x 8 columns
        plt.axis('off')
        plt.title("Label {0} ({1})".format(label, labels.count(label)))
        i += 1
        _ = plt.imshow(image)
    plt.show()

display_images_and_labels(images32, labels)
```

```python
for image in images32[:5]:
    print("shape: {0}, min: {1}, max: {2}".format(image.shape, image.min(), image.max()))
labels_a = np.array(labels)
images_a = np.array(images32)
print("labels: ", labels_a.shape, "\nimages: ", images_a.shape)
```

```python

graph = tf.Graph()

with graph.as_default():
    # Gi ch cho u vo v nhn.
    images_ph = tf.placeholder(tf.float32, [None, 32, 32, 3])
    labels_ph = tf.placeholder(tf.int32, [None])

    # Lm phng u vo t: [Khng, chiu cao, chiu rng, knh]
    # To: [Khng c, chiu cao * chiu rng * knh] == [Khng, 3072]
    images_flat = tf.contrib.layers.flatten(images_ph)

    # Lp kt ni y .
    # To cc bn ghi kch thc [Khng, 62]
    logits = tf.contrib.layers.fully_connected(images_flat, 62, tf.nn.relu)

    # Chuyn i nht k thnh ch mc nhn (int).
    # Hnh dng [Khng], l vect 1D c  di == batch_size.
    predicted_labels = tf.argmax(logits, 1)

    # Xc nh hm mt.
    # Entropy cho l mt la chn tt  phn loi.
    loss = tf.reduce_mean(tf.nn.sparse_softmax_cross_entropy_with_logits(logits=logits, labels=labels_ph))

    # To o to op.
    train = tf.train.AdamOptimizer(learning_rate=0.001).minimize(loss)

    # V cui cng, mt op khi to  thc thi trc khi o to.
    init = tf.global_variables_initializer()

print("images_flat: ", images_flat)
print("logits: ", logits)
print("loss: ", loss)
print("predicted_labels: ", predicted_labels)
```

```python

session = tf.Session(graph=graph)

_ = session.run([init])


for i in range(101):
    _, loss_value = session.run([train, loss], 
                                feed_dict={images_ph: images_a, labels_ph: labels_a})
    if i % 100 == 0:
        print("Loss: ", loss_value)

sample_indexes = random.sample(range(len(images32)), 10)
sample_images = [images32[i] for i in sample_indexes]
sample_labels = [labels[i] for i in sample_indexes]

predicted = session.run([predicted_labels], 
                        feed_dict={images_ph: sample_images})[0]
print(sample_labels)
print(predicted)

```

```python
num_classes = 3 

con = tf.confusion_matrix(labels=sample_labels, predictions=predicted)
sess = tf.Session()
with sess.as_default():
        print(sess.run(con))
```

```python

fig = plt.figure(figsize=(10, 10))
for i in range(len(sample_images)):
    truth = sample_labels[i]
    prediction = predicted[i]
    plt.subplot(5, 2,1+i)
    plt.axis('off')
    color='green' if truth == prediction else 'red'
    plt.text(40, 10, "Truth:        {0}\nPrediction: {1}".format(truth, prediction), 
             fontsize=12, color=color)
    plt.imshow(sample_images[i])


test_images, test_labels = load_data(test_data_dir)

test_images32 = [skimage.transform.resize(image, (32, 32), mode='constant')
                 for image in test_images]
display_images_and_labels(test_images32, test_labels)
```

```python

predicted = session.run([predicted_labels], 
                        feed_dict={images_ph: test_images32})[0]
match_count = sum([int(y == y_) for y, y_ in zip(test_labels, predicted)])
accuracy = match_count / len(test_labels)
print("Accuracy: {:.3f}".format(accuracy))

```

```python

session.close()
print("session close")
```
